```python
##Lectura de archivos


import pandas as pd
import matplotlib.pyplot as plt
#-----Matematicas y quimica-------
archivo_matematicas = '/Users/janeth/Documents/ASISTENCIA GEA MATEMÁTICAS Y QUÍMICA 2024-2S.xlsx'
#------Lectoescritura e ingles-----
archivo_lecto = '/Users/janeth/Documents/ASISTENCIA GEA LECTOESCRITURA E INGLÉS 2024-2S.xlsx'
#------Derecho---------------------
archivo_derecho = '/Users/janeth/Documents/ASISTENCIA GEA PAET 2024-2S.xlsx'

#-----Matematicas-----
df_mate = pd.read_excel(archivo_matematicas, sheet_name='Matemáticas')
conteo_tutores_mate = df_mate['Tutor'].value_counts()
df_mate = df_mate[df_mate['Identificación '].notna()]

#----Quimica---
df_quimica = pd.read_excel(archivo_matematicas, sheet_name='Química')
conteo_tutores_quimica = df_quimica['Tutor'].value_counts()
df_quimica = df_quimica[df_quimica['Identificación '].notna()]

#----Lectoescritura---
df_lecto = pd.read_excel(archivo_lecto, sheet_name='Lectoescritura')
conteo_tutores_lecto = df_lecto['Tutor'].value_counts()
df_lecto = df_lecto[df_lecto['Identificación '].notna()]

#---Derecho---
df_derecho = pd.read_excel(archivo_derecho, sheet_name='Derecho')
conteo_tutores_derecho = df_derecho['Tutor'].value_counts()
df_derecho = df_derecho[df_derecho['Identificación '].notna()]

#----Inglés---
df_ingles = pd.read_excel(archivo_lecto, sheet_name='Inglés')
conteo_tutores_lecto = df_ingles['Tutor'].value_counts()
df_ingles = df_ingles[df_ingles['Identificación '].notna()]

#----- Total----

df_all=pd.concat([df_ingles, df_lecto,df_quimica, df_mate ], ignore_index=True)
df_all.head()
```

```python
###Total asistencias



asistencias_mate = len(df_mate)
asistencias_quimica = len(df_quimica)
asistencias_lecto = len(df_lecto)
asistencias_ingles = len(df_ingles)
asistencias_derecho = len(df_derecho)

total_asistencias = asistencias_mate + asistencias_quimica + asistencias_lecto + asistencias_ingles + asistencias_derecho
#----------------------------------------------------------
porcentaje_mate = (asistencias_mate / total_asistencias) * 100
porcentaje_quimica = (asistencias_quimica / total_asistencias) * 100
porcentaje_lecto = (asistencias_lecto / total_asistencias) * 100
porcentaje_ingles = (asistencias_ingles / total_asistencias) * 100
porcentaje_derecho = (asistencias_derecho / total_asistencias) *100
#-----------------------------------------------------------------------
asistencias_reales_mate = df_mate['Identificación '].nunique()
asistencias_reales_quimica = df_quimica['Identificación '].nunique()
asistencias_reales_lecto = df_lecto['Identificación '].nunique()
asistencias_reales_ingles = df_ingles['Identificación '].nunique()
asistencias_reales_derecho = df_derecho['Identificación '].nunique()
#-----------------------------------------
df_combined = pd.concat([df_mate, df_quimica, df_lecto, df_ingles, df_derecho])
total_real = df_combined['Identificación '].nunique()

#---------------------------------------------------------------------------
data = {
    'Asignatura': ['Matemáticas', 'Lectoescritura', 'Inglés', 'Química','Derecho'],
    'Asistencias': [asistencias_mate, asistencias_lecto, asistencias_ingles, asistencias_quimica, asistencias_derecho],
    '% Asistencias': [porcentaje_mate, porcentaje_lecto, porcentaje_ingles, porcentaje_quimica, porcentaje_derecho]
}

df_resultados = pd.DataFrame(data)
df_resultados['Asistencias reales'] = [asistencias_reales_mate, asistencias_reales_lecto, asistencias_reales_ingles, asistencias_reales_quimica , asistencias_reales_derecho]
df_resultados['%Asistencias en el semestre'] = df_resultados['Asistencias'] / total_asistencias * 100

total = {
    'Asignatura': 'Total',
    'Asistencias': total_asistencias,
    '% Asistencias': 100,
    'Asistencias reales': total_real,
    '%Asistencias en el semestre': (df_resultados['Asistencias reales'].sum() / total_asistencias) * 100
}

df_total = pd.DataFrame([total])
df_resultados = pd.concat([df_resultados, df_total], ignore_index=True)
output_path = '/Users/janeth/Documents/Estadisticas/resultados_asistencias.xlsx'
df_resultados.to_excel(output_path, index=False)
df_resultados
```

```python
#  asistencias y porcentajes por asignatura
def calcular_asistencias_por_asignatura(df):
    asistencias_por_asignatura = df['Asignatura'].value_counts()
    total_asistencias_asignaturas = asistencias_por_asignatura.sum()
    porcentaje_asistencias = (asistencias_por_asignatura / total_asistencias_asignaturas) * 100
    reales=  df.groupby('Asignatura')['Identificación '].nunique()
    df_asistencias_asignaturas = pd.DataFrame({        'Asignatura': asistencias_por_asignatura.index,
        'Asistencias': asistencias_por_asignatura.values,
        '% Asistencias': porcentaje_asistencias.values,
        'Estudiantes atendidos':reales.reindex(asistencias_por_asignatura.index).values
    
    })
    
    return df_asistencias_asignaturas.reset_index(drop=True)

df_asistencias_mate = calcular_asistencias_por_asignatura(df_mate)
df_asistencias_quimica = calcular_asistencias_por_asignatura(df_quimica)
df_asistencias_lecto = calcular_asistencias_por_asignatura(df_lecto)
df_asistencias_ingles = calcular_asistencias_por_asignatura(df_ingles)
df_asistencias_derecho = calcular_asistencias_por_asignatura(df_derecho)

print("Asistencias por asignatura (Matemáticas):")
display(df_asistencias_mate)

print("Asistencias por asignatura (Química):")
display(df_asistencias_quimica)

print("Asistencias por asignatura (Lectoescritura):")
display(df_asistencias_lecto)

print("Asistencias por asignatura (Inglés):")
display(df_asistencias_ingles)

print('Asistencia por asignatura (Derecho)')
display(df_asistencias_derecho)

df_asistencias_mate.to_excel('/Users/janeth/Documents/Estadisticas/asistenciamate.xlsx', index=False)
df_asistencias_quimica.to_excel('/Users/janeth/Documents/Estadisticas/asistenciaquimica.xlsx', index=False)
df_asistencias_lecto.to_excel('/Users/janeth/Documents/Estadisticas/asistencialecto.xlsx', index=False)
df_asistencias_ingles.to_excel('/Users/janeth/Documents/Estadisticas/asistenciaingles.xlsx', index=False)
df_asistencias_derecho.to_excel('/Users/janeth/Documents/Estadisticas/asistenciaderecho.xlsx', index=False)
```

```python

import unicodedata
def limpiar_nombre(nombre):
    if isinstance(nombre, str):
        nombre = nombre.strip().title()  
        nombre = unicodedata.normalize("NFKC", nombre)
        return nombre
    return nombre  
for df in [df_lecto, df_mate, df_quimica, df_ingles]:
    df['Tutor'] = df['Tutor'].apply(limpiar_nombre)
def contar_asistencias(df):
    return df['Tutor'].value_counts()

conteo_df1 = contar_asistencias(df_lecto)
conteo_df2 = contar_asistencias(df_mate)
conteo_df3 = contar_asistencias(df_quimica)
conteo_df4 = contar_asistencias(df_ingles)
conteo_df5 = contar_asistencias(df_derecho)
conteo_total = pd.concat([conteo_df1, conteo_df2, conteo_df3, conteo_df4, conteo_df5], axis=1).fillna(0)
conteo_total['Total'] = conteo_total.sum(axis=1)
conteo_total = conteo_total.sort_values(by='Total', ascending=False)

plt.figure(figsize=(10, 6))
bars = plt.barh(conteo_total.index, conteo_total['Total'], color='orange')
plt.title('Conteo de Asistencias por Tutor')
plt.xlabel('Número de Asistencias')
plt.ylabel('Tutor')

for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{int(bar.get_width())}', 
             va='center', ha='left', color='black', fontweight='bold')
total_tutorias = conteo_total['Total'].sum()
plt.gca().text(0.95, 0.95, f'Total de Tutorías: {total_tutorias}', 
               transform=plt.gca().transAxes, ha='right', va='top', 
               fontsize=10, color='black', fontweight='bold')

plt.show()
```

```python
##Analisis


def distribucion_por_fecha(df):
    df['Fecha'] = pd.to_datetime(df['Fecha : Recordar (DD/MM/AAAA)'], format='%d/%m/%Y')
    df['Mes'] = df['Fecha'].dt.month
    asistencias_por_mes = df['Mes'].value_counts().sort_index()
    asistencias_por_semana = df['Semana académica'].value_counts().sort_index()
    asistencias_por_dia = df['Día de la semana'].value_counts().sort_index()
    return asistencias_por_mes, asistencias_por_semana, asistencias_por_dia
def distribucion_por_jornada(df):
    asistencias_por_jornada = df['Jornada'].value_counts().sort_index()
    jornada_mas_asistencias = asistencias_por_jornada.idxmax()
    return asistencias_por_jornada, jornada_mas_asistencias
def distribucion_por_programa(df):
    asistencias_por_programa = df['Programa curricular'].astype(str).value_counts().sort_index()
    return asistencias_por_programa
def distribucion_por_tutor(df):
    asistencias_por_tutor = df['Tutor'].value_counts().sort_index()
    return asistencias_por_tutor
def distribucion_por_asignatura(df):
    asistencias_por_asignatura = df['Asignatura'].value_counts().sort_index()
    return asistencias_por_asignatura
def motivos_de_consulta(df):
    motivos = df['Motivo de la consulta'].value_counts().sort_index()
    return motivos
def calcular_porcentajes(series):
    total = series.sum()
    porcentajes = (series / total) * 100
    return porcentajes
dataframes = {
    'lectoescritura': df_lecto,
    'matemáticas': df_mate,
    'química': df_quimica,
    'inglés': df_ingles,
    'derecho': df_derecho
}
resultados = {
    'fecha': {},
    'jornada': {},
    'programa': {},
    'tutor': {},
    'asignatura': {},
    'motivos': {},
    'programas_especiales': {}
}
for area, df in dataframes.items():
    asistencias_por_mes, asistencias_por_semana, asistencias_por_dia = distribucion_por_fecha(df)
    resultados['fecha'][area] = {
        'mes': asistencias_por_mes,
        'semana': asistencias_por_semana,
        'dia': asistencias_por_dia
    }
    asistencias_por_jornada, jornada_mas_asistencias = distribucion_por_jornada(df)
    resultados['jornada'][area] = {
        'asistencias': asistencias_por_jornada,
        'jornada_mas_asistencias': jornada_mas_asistencias
    }
    asistencias_por_programa = distribucion_por_programa(df)
    resultados['programa'][area] = asistencias_por_programa
    asistencias_por_tutor = distribucion_por_tutor(df)
    resultados['tutor'][area] = asistencias_por_tutor
    asistencias_por_asignatura = distribucion_por_asignatura(df)
    resultados['asignatura'][area] = asistencias_por_asignatura
    motivos = motivos_de_consulta(df)
    resultados['motivos'][area] = motivos
for area, data in resultados['fecha'].items():
    porcentaje_dia = calcular_porcentajes(data['dia'])
    print(f"Porcentaje de asistencias por día de la semana en {area}:\n", porcentaje_dia)
for area, data in resultados['motivos'].items():
    porcentaje_motivos = calcular_porcentajes(data)
    print(f"Porcentaje de motivos de consulta en {area}:\n", porcentaje_motivos)
for area, data in resultados['jornada'].items():
    print(f"Asistencias por jornada en {area}:\n", data['asistencias'])
    print(f"Jornada con más asistencias en {area}: {data['jornada_mas_asistencias']}")
for area, data in resultados['programa'].items():
    print(f"Asistencias por programa curricular en {area}:\n", data)
for area, data in resultados['tutor'].items():
    print(f"Asistencias por tutor en {area}:\n", data)
for area, data in resultados['asignatura'].items():
    print(f"Asistencias por asignatura en {area}:\n", data)
```
